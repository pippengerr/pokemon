```python
# %%pycodestyle


# pip install termcolor, pyfiglet, colorama
from termcolor import cprint
from pyfiglet import figlet_format
from colorama import init
import random
import time
import sys
init(strip=not sys.stdout.isatty())


class Pokemon:
    def __init__(self, name, Level, Type, HP, Attack, Defense, Speed,
                 Experience, Moves):
        self.name = name
        self.type = Type
        self.hp = HP
        self.attack = Attack
        self.defense = Defense
        self.speed = Speed
        self.exp_given = Experience
        self.moves = Moves
        self.level = Level

    def calculate_dmg(self, enemy, movenum):
        """Calculates the damage dealt to the defending pokemon (enemy)
        by the attacking pokemon (self). movenum refers to the move in
        the move list of the attacking pokemon."""
        level = self.level  # level of attacking pokemon
        move = self.moves[movenum - 1]  # move selected
        power = MOVES_DICTIONARY[move]['power']  # power of move
        a = self.attack  # attack of attacking pokemon
        speed = self.speed  # speed of attacking pokemon
        d = enemy.defense  # defense of defending pokemon

        # random percent between 85% and 100% inclusive
        randpercent = random.randint(85, 100)/100.

        # Type effectiveness
        enemytype = enemy.type  # Type of defending pokemon
        effectiveness = 1
        for x in enemytype:
            if x in MOVES_DICTIONARY[move]['not very effective against']:
                effectiveness *= .5
            elif x in MOVES_DICTIONARY[move]['super effective against']:
                effectiveness *= 2.
            else:
                effectiveness *= 1.

        # Chance of critical hit (speed/512)
        critical = 1
        if random.randint(0, 511) < speed:
            critical = 2

        # Calculating the modifier
        modifier = randpercent * effectiveness * critical

        dmg = (((2.*level/5.+2)*power*a/d)/50.)*modifier
        return round(dmg, 2), critical

    def takedmg(self, dmgtaken):
        """Decreases the pokemon's hp based on how much damage
        they have taken."""
        self.hp = max(self.hp - dmgtaken, 0)

    def __str__(self):
        return (f'{self.name} / HP: {round(self.hp, 2)} / Level: '
                f'{self.level} / Moves: {self.moves}.')
```

```python
# %%pycodestyle


class Pokegame:
    def choose_pokemon(self, pokemonlist, startexp):
        """Lets the user input a pokemon name and level to choose
        their pokemon. Checks to see if the input pokemon is part of the
        available pokemon list and that the level is an integer."""
        level = round(startexp ** (1/3))
        while True:
            userinput = str(input('Choose your Pokemon: ').title())
            if userinput in pokemonlist:
                userpoke = Pokemon(userinput, level, **CHARACTERS[userinput])
                return userpoke
            elif userinput not in pokemonlist:
                print('Pokemon not recognized or is still locked...(e.g.'
                      ' "Squirtle"  or "squirtle" is accepted but'
                      '"sqwertle" is not) \n')

    def generate_enemy(self):
        """Generates a random enemy from the list of 14 pokémon with levels
        ranging from 1 to 2."""
        enemypokename = random.choice(list(CHARACTERS.keys()))
        enemylevel = random.randint(1, 2)
        enemypoke = Pokemon(enemypokename, 0,
                            **CHARACTERS[enemypokename])
        enemypoke.level = enemylevel
        return enemypoke

    def choose_move(self, playerpoke, enemy):
        """Player gets to choose a move from the options their Pokémon
        can utilize."""
        movelist = f'{playerpoke.name} has these moves: \n'
        i = 1
        for x in playerpoke.moves:
            movelist += f'{i}) {x}\n'
            i += 1
        print(movelist)
        while True:
            try:
                playerfight = int(input(
                    'Which move do you want to choose? (Please'
                    ' enter the number. ex: 1, 2, or 3) '))
                if playerfight <= 0:
                    raise ValueError
                dmgdealt, critical = playerpoke.calculate_dmg(
                    enemy, playerfight)
            except ValueError:
                print('Please enter a valid number...')
            except IndexError:
                print('Please enter a valid number...')
            else:
                break
        return playerpoke.moves[playerfight-1], dmgdealt, critical

    def enemy_move(self, userpoke, enemy):
        """Chooses the most damaging move the enemy can use against userpoke.
        Calculates the damage for each move and saves them to a dictionary.
        Chooses the largest value and key as the most damaging move.
        Returns the move and max damage."""
        movedict = {}
        for movenum in range(len(enemy.moves)):
            movedict[enemy.moves[movenum]], critical = enemy.calculate_dmg(
                userpoke, movenum)
        maxdmg = max(movedict.values())
        maxdmg_move = [x for x, y in movedict.items() if y == maxdmg]
        return (maxdmg_move[0], maxdmg, critical)

    def addtocollection(self, initcollection, newpokemon):
        """Checks if the newpokemon is in the initcollection so that the user
        can play as the pokemon it has just defeated.
        If not in initcollection, newpokemon is added to the collection."""
        if newpokemon not in initcollection:
            initcollection.append(newpokemon)
            print(f'{newpokemon} has been added to your collection!\n')
        return initcollection

    def make_availpokelist(self, pokelist):
        """Makes a string of the available pokemon for the player to
        choose."""
        pokemonavail = 'List of available Pokemon: \n'
        for x in pokelist:
            temp = Pokemon(x, 0, **CHARACTERS[x])
            pokemonavail += f'{x} / HP: {temp.hp} / Moves: {temp.moves} \n'
        return pokemonavail

    def banner(self, text, ch='=', length=78):
        spaced_text = ' %s ' % text
        banner = spaced_text.center(length, ch)
        return banner

    def main(self):
        '''Game initialization. Following that, a banner and introduction
        are printed, timed so that the user is not overwhelmed by text.'''
        game = Pokegame()
        # totalexp is the initial exp points your pokemon starts with
        totalexp = 1000000

        print(game.banner('Welcome to the Pokémon World'))
        print('\n')
        time.sleep(2)
        print(game.banner('^^^^^^^^^^^^^^^^'))

        cprint(figlet_format('Welcome ', font='starwars'),
               'yellow', 'on_red', attrs=['bold'])
        print(game.banner('^^^^^^^^^^^^^^^^'))
        print('     Hello there! Welcome to the world of pokémon! My name'
              ' is Oak! People call me the pokémon Prof! This world is '
              'inhabited by creatures called pokémon! For some people, '
              'pokémon are pets. Others use them for fights. Myself...I '
              'study pokémon as a profession.\n')
        time.sleep(4)
        print("     Pokémon—short for pocket monsters—is the name of a "
              'game (really several) involving creatures called “pokémon”'
              ' and “trainers.” Within the narrative of the game, the '
              'pokémon trainer catches pokémon in little holding containers'
              ' (called “pokeballs”), and then uses those pokémon to fight'
              ' other pokémon. On its surface, the fights have two reasons:'
              ' 1) to weaken and capture wild pokémon, and 2) to defeat '
              'other pokémon trainers. The pokémon themselves are various,'
              ' having different appearances, names, powers, potentials, '
              'weaknesses, and personalities.\n')
        time.sleep(4)
        print("     Any good trainer knows that you can't learn unless you"
              ' play the game. Being a good trainer is a long and '
              'complicated process which involves working with the pokémon'
              ' you have in order to accumulate experience and catch more '
              'pokémon. The more one loses oneself in the pursuit of '
              'pokémon, the greater the rewards. Pokémon evolve, they grow'
              ' stronger, and thus lead the trainer to greater captures and'
              ' greater pursuits!\n')
        time.sleep(4)
        print("     Now go ahead and start your adventure!!!!\n\n")
        time.sleep(3)

        # Initial collection of pokemon
        pokelist = ['Bulbasaur', 'Squirtle',
                    'Charmander', 'Pikachu']
        print(game.make_availpokelist(pokelist), '\n')
        time.sleep(1)

        # User chooses their starter pokemon
        userpokemon = game.choose_pokemon(pokelist, totalexp)
        print(f'Your Pokemon: {userpokemon} \n')
        time.sleep(2)

        while userpokemon.hp > 0:
            # Keeps battles looping.
            # Enemy chooses a random pokemon
            enemypokemon = game.generate_enemy()
            print(f'A wild Pokemon has appeared! \n{enemypokemon.name} '
                  f'/ HP: {enemypokemon.hp} / Level: {enemypokemon.level}.\n')
            time.sleep(2)

            turn = 1
            while userpokemon.hp > 0 and enemypokemon.hp > 0:
                # Keeps turns looping
                print(f'Turn {turn}\n')
                time.sleep(2)

                # Player chooses their move
                playermovename, playerdmg, playercrit = game.choose_move(
                    userpokemon, enemypokemon)
                compmovename, compdmg, enemycrit = game.enemy_move(
                    userpokemon, enemypokemon)
                turn += 1

                # Check speed/order of player and enemy pokemon moves
                userspeed = userpokemon.speed
                enemyspeed = enemypokemon.speed
                if userspeed > enemyspeed:
                    print(f"You attack first")
                    time.sleep(1)
                    print(f"Your {userpokemon.name} used {playermovename}")
                    time.sleep(1)
                    if playercrit == 2:
                        print(f'It was a critical hit!')
                        time.sleep(1)
                    print(f"Your move dealt {playerdmg} points of damage")
                    time.sleep(1)
                    enemypokemon.takedmg(playerdmg)
                    print(f"The {enemypokemon.name}'s HP drops to"
                          f" {round(enemypokemon.hp, 2)}\n")
                    time.sleep(2)
                    # If enemy pokemon is still alive
                    if enemypokemon.hp > 0:
                        print(f"The enemy attacks you! ")
                        time.sleep(1)
                        print(f"The enemy used {compmovename}")
                        time.sleep(1)
                        if enemycrit == 2:
                            print(f'It was a critical hit!')
                            time.sleep(1)
                        print(f"Their move dealt {compdmg} "
                              f"points of damage \n")
                        time.sleep(1)
                        userpokemon.takedmg(compdmg)
                        print(userpokemon)
                        print(f'{enemypokemon.name} / HP: '
                              f'{round(enemypokemon.hp, 2)} / '
                              f'Level: {enemypokemon.level}\n')

                elif userspeed < enemyspeed:
                    print(f"The enemy attacks first ")
                    time.sleep(1)
                    print(f"The enemy used {compmovename}")
                    time.sleep(1)
                    if enemycrit == 2:
                        print(f'It was a critical hit!')
                        time.sleep(1)
                    print(f"Their move dealt {compdmg} points of damage")
                    time.sleep(1)
                    userpokemon.takedmg(compdmg)
                    print(f"Your {userpokemon.name}'s HP drops to "
                          f"{round(userpokemon.hp, 2)}\n")
                    time.sleep(2)
                    # If your pokemon is still alive
                    if userpokemon.hp > 0:
                        print(f"It's your turn now!")
                        time.sleep(1)
                        print(f"Your {userpokemon.name} used {playermovename}")
                        time.sleep(1)
                        if playercrit == 2:
                            print(f'It was a critical hit!')
                            time.sleep(1)
                        print(f"Your move dealt {playerdmg} points of "
                              f"damage \n")
                        time.sleep(1)
                        enemypokemon.takedmg(playerdmg)
                        print(userpokemon)
                        time.sleep(1)
                        print(f'{enemypokemon.name} / HP: '
                              f'{round(enemypokemon.hp, 2)} / '
                              f'Level: {enemypokemon.level}\n')
                        time.sleep(1)

                elif userspeed == enemyspeed:
                    attack = random.randint(1, 2)
                    if attack == 1:
                        print(f"The enemy attacks first ")
                        time.sleep(1)
                        print(f"The enemy used {compmovename}")
                        time.sleep(1)
                        if enemycrit == 2:
                            print(f'It was a critical hit!')
                            time.sleep(1)
                        print(f"Their move dealt {compdmg} points of damage")
                        time.sleep(1)
                        userpokemon.takedmg(compdmg)
                        print(f"Your {userpokemon.name}'s HP drops to"
                              f" {round(userpokemon.hp, 2)}\n")
                        time.sleep(2)
                        # If your pokemon is still alive
                        if userpokemon.hp > 0:
                            print(f"It's your turn now!")
                            time.sleep(1)
                            print(f"Your {userpokemon.name} used"
                                  f" {playermovename}")
                            time.sleep(1)
                            if playercrit == 2:
                                print(f'It was a critical hit!')
                                time.sleep(1)
                            print(f"Your move dealt {playerdmg} points "
                                  f"of damage \n")
                            time.sleep(1)
                            enemypokemon.takedmg(playerdmg)
                            print(userpokemon)
                            time.sleep(1)
                            print(f'{enemypokemon.name} / HP: '
                                  f'{round(enemypokemon.hp, 2)} / '
                                  f'Level: {enemypokemon.level}\n')
                            time.sleep(1)
                    if attack == 2:
                        print(f"You attack first")
                        time.sleep(1)
                        print(f"Your {userpokemon.name} used "
                              f"{playermovename}")
                        time.sleep(1)
                        if playercrit == 2:
                            print(f'It was a critical hit!')
                            time.sleep(1)
                        print(f"Your move dealt {playerdmg} points of damage")
                        time.sleep(1)
                        enemypokemon.takedmg(playerdmg)
                        print(f"The {enemypokemon.name}'s HP drops to"
                              f" {round(enemypokemon.hp, 2)}\n")
                        time.sleep(2)
                        # If enemy pokemon is still alive
                        if enemypokemon.hp > 0:
                            print(f"The enemy attacks you! ")
                            time.sleep(1)
                            print(f"The enemy used {compmovename}")
                            time.sleep(1)
                            if enemycrit == 2:
                                print(f'It was a critical hit!')
                                time.sleep(1)
                            print(f"Their move dealt {compdmg} points "
                                  f"of damage \n")
                            time.sleep(1)
                            userpokemon.takedmg(compdmg)
                            print(userpokemon)
                            print(f'{enemypokemon.name} / HP: '
                                  f'{round(enemypokemon.hp, 2)} /'
                                  f' Level: {enemypokemon.level}\n')

            # When enemy faints, add it to the pokelist
            if enemypokemon.hp <= 0 and userpokemon.hp > 0:

                # Adds exp to total
                totalexp += enemypokemon.exp_given
                print(f'Enemy {enemypokemon.name} has been defeated! You '
                      f'gained {enemypokemon.exp_given} EXP!\n')
                time.sleep(1)

                # Add it to the collection. Append to pokelist
                pokelist = game.addtocollection(pokelist, enemypokemon.name)
                time.sleep(1)

                # Ask player for their next pokemon
                print('Get ready for your next battle!\n')
                time.sleep(2)
                print(game.make_availpokelist(pokelist), '\n')
                userpokemon = game.choose_pokemon(pokelist, totalexp)
                print(f'Your Pokemon: {userpokemon} \n')
                time.sleep(2)

        # When player's pokemon faints, game over!
        if enemypokemon.hp > 0 and userpokemon.hp <= 0:
            print('Your Pokemon has fainted. Game over!')


# run the main program
if __name__ == '__main__':
    Pokegame().main()
```
